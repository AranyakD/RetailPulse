```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv("../data/processed/cleaned_data.csv")
```

```python
df.groupby("Month") ["TotalPrice"].sum().sort_index().plot()
plt.title("Monthly Sales")
plt.show()
```

```python
df.groupby("Country")["TotalPrice"].sum().sort_values(ascending=False).head(10).plot(kind="bar")
plt.title("Top Countries")
plt.show()
```

```python
np.log1p(df["TotalPrice"]).hist(bins=50)
plt.title("Revenue Distribution")
plt.show()
```

```python
plt.scatter(df["Quantity"], df["TotalPrice"], alpha=0.3)
plt.xlabel("Quantity")
plt.ylabel("Total Price")
plt.title("Quantity vs Revenue")
plt.show()
```
